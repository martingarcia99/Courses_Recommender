```python
import re
import os
import nltk
import string
import pandas as pd
from bs4 import BeautifulSoup
from dotenv import load_dotenv
from pymongo import MongoClient
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer 
from nltk.tokenize import RegexpTokenizer 
from nltk.stem import WordNetLemmatizer 
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
ps = PorterStemmer() 
lemmatizer = WordNetLemmatizer() 
stop_words = set(stopwords.words('english')) 
tokenizer = RegexpTokenizer(r'\w+')
df = pd.DataFrame(list(lectures.find()))
data=df.drop(columns = ['_id', 'avg_rating', 'assigned_people', 'language', 'course_format', 'study_courses', 'positive_comments', 'negative_comments', 'lecture_id', 'semester', 'comments','pre_qualifications'])

for col in data.columns:
    print(col)
    
def remove_html(text):
    soup = BeautifulSoup(text, 'lxml')
    html_free = soup.get_text()
    return html_free

def remove_punctuation(text):
    no_punct = ''.join([c for c in text if c not in string.punctuation])
    return no_punct

def remove_stop_word(text):
    no_stop_word = [c for c in text if c not in stop_words]
    return no_stop_word

def word_lemmatizer(text):
    lem_text = [lemmatizer.lemmatize(i) for i in text]
    return lem_text

data['description'] = data['description'].apply(lambda x: remove_html(x))
data['description'] = data['description'].apply(lambda x: remove_punctuation(x))
data['description'] = data['description'].apply(lambda x: tokenizer.tokenize(x.lower()))
data['description'] = data['description'].apply(lambda x: remove_stop_word(x))
data['description'] = data['description'].apply(lambda x: word_lemmatizer(x))
data['description'][0]
```
